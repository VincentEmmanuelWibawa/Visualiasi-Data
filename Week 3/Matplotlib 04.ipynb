# Sample Data

```python
mu, sigma = 100, 15 

x = mu + sigma * np.random.randn(10000)
x
```

```python
x.shape
```

# Histogram dengan pyplot Style

```python
plt.hist(x,
         bins=50,
         facecolor='g',
         alpha=0.75)

plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Contoh Histogram')

plt.text(45, 500, r'$\mu=100, \ \sigma=15$')
plt.grid()

plt.show()
```
